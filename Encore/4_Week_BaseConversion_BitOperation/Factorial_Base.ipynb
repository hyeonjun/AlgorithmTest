# 팩토리얼 진법

### - 문제설명

상근이는 보통 사람들이 사는 것과는 조금 다른 삶을 사는 사람이다. 상근이는 이런 사람들의 시선이 부담스럽기 때문에, 자신만의 숫자를 개발하기로 했다. 바로 그 이름은 팩토리얼 진법이다. 팩토리얼 진법은 각 자리에 올 수 있는 숫자는 0부터 9까지로 10진법과 거의 비슷하다. 하지만, 읽는 법은 조금 다르다. 팩토리얼 진법에서는 i번 자리의 값을 ai×i!로 계산한다. 즉, 팩토리얼 진법에서 719는 10진법에서 53과 같다. 그 이유는 7×3! + 1×2! + 9×1! = 53이기 때문이다.

팩토리얼 진법으로 작성한 숫자가 주어졌을 때, 10진법으로 읽은 값을 구하는 프로그램을 작성하시오. 

### - 입력

입력은 여러 개의 테스트 케이스로 이루어져 있다. 각 테스트 케이스는 한 줄로 이루어져 있고, 길이가 최대 5자리인 팩토리얼 진법 숫자가 주어진다. 입력의 마지막 줄에는 0이 하나 주어진다.

### - 출력

각 테스트 케이스에 대해서, 입력으로 주어진 팩토리얼 진법 숫자를 10진법으로 읽은 값을 출력한다.

#### 풀이

다른 진법과 구하는 방법은 같지만 팩토리얼이라는 점만 생각하면 될듯하다
그래서 팩토리얼 연산하는 방법만 구한다면 아무 문제 없을 것 같다.

파이썬의 Math 라이브러리의 factorial 함수가 있지만 나는 직접 구현해서 풀어볼 것이다.

팩토리얼은 N! = N * (N-1) * (N-2) .... * 1 이기 때문에 가장 쉽고 간단하게 풀 수 있는 방식이 재귀일 것이다. 그래서 factory라는 함수를 만들어 재귀를 시켜볼 것이다. 여기서 중요한 점은 1에서 끝나고 1! = 1이므로 factory 함수가 전달받은 인자값이 1이면 1을 리턴시켜주어야 모든 연산이 끝나고 돌아와서 다시 리턴될 때의 값이 N * (N-1) * (N-2) .... * 1 가 될 것이다. 


```python
import sys
input = sys.stdin.readline
def factory(n):
    if n == 1:
        return 1
    return n * factory(n-1)

while True:
    n = input().strip()
    if n == '0':
        break
    answer = 0
    for i in range(len(n)):
        answer += int(n[i]) * factory(len(n)-i)
    print(answer)

```
