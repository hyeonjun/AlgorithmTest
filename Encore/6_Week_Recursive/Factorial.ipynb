# 팩토리얼

### - 문제

0보다 크거나 같은 정수 N이 주어진다. 이때, N!을 출력하는 프로그램을 작성하시오.

### - 입력

첫째 줄에 정수 N(0 ≤ N ≤ 12)이 주어진다.

### - 출력

첫째 줄에 N!을 출력한다.

image.png

```python
# 파이썬 Math의 factorial 사용
from math import factorial
print(factorial(int(input())))
print()
print(factorial(int(input())))
```

image.png

```python
# 재귀 함수
def factorial(n):
    return n * factorial(n-1) if n > 1 else 1
print(factorial(int(input())))
print()
print(factorial(int(input())))
```
