# 이진수 변환

### - 문제

자연수 N이 주어진다. N을 이진수로 바꿔서 출력하는 프로그램을 작성하시오.

### - 입력

첫째 줄에 자연수 N이 주어진다. (1 ≤ N ≤ 100,000,000,000,000)

### - 출력

N을 이진수로 바꿔서 출력한다. 이진수는 0으로 시작하면 안 된다.

image.png

```python
# 파이썬 bin() 함수 사용
print(bin(int(input()))[2:])
```

image.png

```python
# 재귀 함수
def binary(n):
    s = "01"
    q, r = divmod(n, 2)
    return s[r] if q == 0 else binary(q) + s[r]

print(binary(int(input())))
```
