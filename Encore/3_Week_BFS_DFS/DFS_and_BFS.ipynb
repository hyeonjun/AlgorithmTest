# DFS와 BFS

### - 문제설명

그래프를 DFS로 탐색한 결과와 BFS로 탐색한 결과를 출력하는 프로그램을 작성하시오. 단, 방문할 수 있는 정점이 여러 개인 경우에는 정점 번호가 작은 것을 먼저 방문하고, 더 이상 방문할 수 있는 점이 없는 경우 종료한다. 정점 번호는 1번부터 N번까지이다.

### - 입력

첫째 줄에 정점의 개수 N(1 ≤ N ≤ 1,000), 간선의 개수 M(1 ≤ M ≤ 10,000), 탐색을 시작할 정점의 번호 V가 주어진다. 다음 M개의 줄에는 간선이 연결하는 두 정점의 번호가 주어진다. 어떤 두 정점 사이에 여러 개의 간선이 있을 수 있다. 입력으로 주어지는 간선은 양방향이다.

### - 출력

첫째 줄에 DFS를 수행한 결과를, 그 다음 줄에는 BFS를 수행한 결과를 출력한다. V부터 방문된 점을 순서대로 출력하면 된다.

순서대로 탐색하게 된다.

dfs는 dfs함수로 들어오게되면 바로 answer 리스트에 담고 다음 아래 노드가 answer 리스트에 들어있지않으면 다시 아래 쪽 노드로 진입하도록 재귀하였다.

bfs는 visited 리스트와 queue를 사용하였다. visited 리스트를 사용하여 방문 순서를 담았고, 이 리스트로 방문 순서를 담아 답을 출력할 수 있다.



```python
# 테스트 해보기 위해서 solution 함수로 만들어 실행하였다.
def solution():
    answer_dfs = []

    def dfs(start):
        answer_dfs.append(str(start))
        for n in graph[start]:
            if str(n) not in answer_dfs:
                dfs(n)

    def bfs(start):
        visited = [str(start)]
        queue = [start]
        while queue:
            x = queue.pop(0)
            for n in graph[x]:
                if str(n) not in visited:
                    queue.append(n)
                    visited.append(str(n))
        return visited

    n, m, v = map(int, input().split())
    graph = [[] for _ in range(n + 1)]

    for _ in range(m):
        a, b = map(int, input().split())
        graph[a].append(b)
        graph[b].append(a)

    for i in graph:
        i.sort()

    dfs(v)
    return answer_dfs, bfs(v)
```

```python
dfs, bfs = solution()
print(" ".join(dfs))
print(" ".join(bfs))
```
